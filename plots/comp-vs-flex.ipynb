```python
import matplotlib            # top-level import only
import matplotlib.pyplot      # access pyplot via full path
```

```python
# Data
offers = [
    ("Big Tech", 2, 450),
    ("Enterprise", 4, 330),
    ("Late-stage startup", 6, 300),
    ("Growth-stage startup", 7, 260),
    ("Series A startup", 8, 210),
    ("Seed-stage startup", 9, 180),
    ("Boutique consultancy", 10, 150),
]

labels = [o[0] for o in offers]
x_vals = [o[1] for o in offers]
y_vals = [o[2] for o in offers]

# Set up figure and axes
fig, ax = matplotlib.pyplot.subplots(figsize=(8, 4))

# Hide all ticks and tick labels
ax.set_xticks([])
ax.set_yticks([])
ax.tick_params(bottom=False, left=False, labelbottom=False, labelleft=False)

ax.set_ylabel("Compensation")
ax.set_xlabel("Flexibility")

# Remove spines
for spine in ax.spines.values():
    spine.set_visible(False)

# Draw custom axis arrows
#ax.axhline(y=min(y_vals) - 20, xmin=0.02, xmax=0.98, color='black', linewidth=1.0, clip_on=False)
#ax.axvline(x=min(x_vals) - 0.5, ymin=0.02, ymax=0.98, color='black', linewidth=1.0, clip_on=False)
ax.annotate('', xy=(max(x_vals) + 0.5, min(y_vals) - 20), 
            xytext=(min(x_vals) - 0.5, min(y_vals) - 20), 
            arrowprops=dict(arrowstyle='->', lw=1.0))
ax.annotate('', xy=(min(x_vals) - 0.5, max(y_vals) + 40), 
            xytext=(min(x_vals) - 0.5, min(y_vals) - 20), 
            arrowprops=dict(arrowstyle='->', lw=1.0))

# Plot points
ax.scatter(x_vals, y_vals)

# Annotate with smart label positioning
for x, y, label in zip(x_vals, y_vals, labels):
    if x <= 6:
        ax.text(x + 0.2, y , label, ha='left', va='center', fontsize="smaller")
    else:
        ax.text(x - 0.2, y , label, ha='right', va='center', fontsize="smaller")

# Set limits with padding to prevent clipping
ax.set_xlim(min(x_vals) - 1, max(x_vals) + 1)
ax.set_ylim(min(y_vals) - 30, max(y_vals) + 50)
```
